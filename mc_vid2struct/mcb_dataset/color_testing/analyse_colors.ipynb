```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
```

```python
image = Image.open('color_example.png').convert('HSV')
white_threshold = 0.9  # Adjust threshold as needed
non_white_pixels = (np.array(image)[:, :, 1] / 255) > 0.5
hsv_image_filtered = np.array(image)[non_white_pixels]
hue_values = hsv_image_filtered[:, 0] / 255 * 360
plt.figure(figsize=(10,4))
plt.hist(hue_values, bins=360//5, range=(0, 360), color='blue', edgecolor='black')
plt.title('Histogram of Hue Values')
plt.xlabel('Hue')
plt.ylabel('Frequency')
plt.xticks([i*20 for i in range(18)])
plt.show()
```
